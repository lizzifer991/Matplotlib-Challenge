## Observations and Insights 

```python
# Dependencies and Setup
%matplotlib notebook
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_study = pd.merge(mouse_metadata, study_results, how="left", on=["Mouse ID", "Mouse ID"])
# Display the data table for preview
mouse_study
```

```python
# Checking the number of entries.
mouse_study.count()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = mouse_study[mouse_study.duplicated(subset= ["Mouse ID", "Timepoint"])]
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
dup_mouse1=mouse_study.loc[mouse_study["Mouse ID"] =="g989"]
dup_mouse1
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
dup_free_study = mouse_study[mouse_study["Mouse ID"] != "g989"]
dup_free_study
```

```python
# Checking the number of mice in the clean DataFrame.
clean_study= dup_free_study["Mouse ID"].value_counts().count()
clean_study
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
regimen= dup_free_study.groupby(["Drug Regimen"])["Tumor Volume (mm3)"]
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: mean, median, variance, standard deviation, and SEM of the tumor volume. 
mean= regimen.mean()
median= regimen.median()
variance= regimen.var()
st_dev= regimen.std()
sem= regimen.sem()
# Assemble the resulting series into a single summary dataframe.
summary = pd.DataFrame({"Mean": mean,
                                  "Median": median,
                                  "Variance": variance,
                                  "Std. Dev": st_dev,
                                  "SEM": sem})
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary
```

```python
# Using the aggregation method, produce the same summary statistics in a single line
summary_agg = regimen.agg(["mean", "median", "var", "std", "sem"])
summary_agg
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.
regimen_timepoint = dup_free_study[["Drug Regimen", "Timepoint"]]
#Calculate the number of timepoint enteries per regimen
regimen_timepoint= regimen_timepoint.groupby(["Drug Regimen"]).count()
#Create bar plot
regimen_timepoint.plot(kind="bar", figsize=(10,5),rot=0, title = "# of Mesurements per Drug Regimen", xlabel="Drug Regimen", ylabel="# of Timepoints")
```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.
regimen_mouse = dup_free_study[["Mouse ID", "Drug Regimen", "Timepoint"]]
#Calculate the number of timepoint & mice per regimen
regimen_mouse = regimen_mouse.groupby(["Drug Regimen"]).count()
#Create bar plot
regimen_mouse.plot(kind="bar", figsize=(10,5), rot=0, title= "# of Mice & Measurement per Drug", xlabel= "Drug Regimen", ylabel="# of Timepoints") 
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
#Filter by gender
mice_gender=dup_free_study[["Mouse ID", "Sex"]]
#Drop Duplicate mouse ID
mice_gender = mice_gender.drop_duplicates()
#Calculate mice by gender
mice_gender = mice_gender.groupby(["Sex"]).count()
#Create pie chart
mice_gender.plot.pie(y="Mouse ID", figsize=(8,8))
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
mice_gender.groupby(["Sex"]).sum().plot.pie(y="Mouse ID", figsize=(8,8), autopct="%1.1f%%")
```
